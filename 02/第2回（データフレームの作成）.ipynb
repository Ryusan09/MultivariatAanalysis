多変量解析を実施するうえで必要となるデータフレームについて述べる．

次のようにデータを入力するのは，1変量のデータを3つ入力しただけで，多変量データの形式にはなっていないことに注意する必要がある．

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
heights = np.array([172, 164, 165, 164, 166, 182, 156, 166, 169, 169, 172, 166, 182, 175, 172, 171, 160, 150, 159, 168])  # 身長
weights = np.array([57, 46, 52, 52, 62, 78, 52, 61, 67, 65, 60, 50, 85, 49, 60, 56, 50, 53, 54, 59]) # 体重
waists = np.array([81, 70, 71, 75, 86, 107, 69, 99, 83, 109, 95, 87, 94, 67, 95, 88, 87, 71, 80, 69]) # ウエスト（腹囲）
```

Pythonで多変量解析を実施する場合，1変量のデータを合併して，多変量データの形式にする必要がある．そこで使うのがデータフレームの作成である．PandasのDataFrame関数を用いる．

```python
datay = pd.DataFrame({'heights': heights, 'weights': weights, 'waists': waists})
```

これでdatayという名前のデータフレームが作成できる．datayの内容を確認すると，次のようになっていることが分かる．

```python
print(datay)
```

Pandasのscatter_matrix関数を使うと，以下の図に示すような散布図行列を作成することができる．

```python
plt.figure(figsize=(10, 8))
pd.plotting.scatter_matrix(datay, figsize=(10, 8), diagonal='hist', alpha=0.7)
plt.suptitle('ペアプロット(pandas)', fontsize=16)
plt.tight_layout()
plt.show()
```

Pythonで多変量解析を実施するときには，データフレームの形式でデータを解析することが基本となる．

なお，相関係数行列（相関係数を配置した行列）は次のようになる．

```python
print("相関係数行列:")
print(datay.corr())
```
