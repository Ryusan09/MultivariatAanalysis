Pythonで四分位数を計算するには、主に以下の3つの方法があります

### numpy ライブラリを使用する

```python
import numpy as np

data = np.array([9, 13, 14, 21, 24, 30, 40, 44, 45, 45, 48])
q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)  # 中央値
q3 = np.percentile(data, 75)

print(f"第1四分位数: {q1}")
print(f"第2四分位数: {q2}")  # 中央値
print(f"第3四分位数: {q3}")
```

### pandas ライブラリを使用する

```python
import pandas as pd

data = pd.Series([9, 13, 14, 21, 24, 30, 40, 44, 45, 45, 48])
q1 = data.quantile(0.25)
q2 = data.quantile(0.5)  # 中央値
q3 = data.quantile(0.75)

print(f"第1四分位数: {q1}")
print(f"第2四分位数: {q2}")  # 中央値
print(f"第3四分位数: {q3}")
```

### statistics モジュールを使用する

```python
import statistics

data = [9, 13, 14, 21, 24, 30, 40, 44, 45, 45, 48]
q1, q2, q3 = statistics.quantiles(data, method='inclusive')

print(f"第1四分位数: {q1}")
print(f"第2四分位数: {q2}")  # 中央値
print(f"第3四分位数: {q3}")
```
