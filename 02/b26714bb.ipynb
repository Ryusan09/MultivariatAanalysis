## その１

```python
import pandas as pd
import numpy as np

d3_2 = "./data3_2.csv"
# d4_6 = "./data4_6.csv"
# CSVファイルのデータをデータフレームに読み込む
df = pd.read_csv(
    # 同じフォルダー内のCSVファイルを読み込む
    d3_2,
    # 見出し列のインデックス0を指定
    index_col=0,
)
# データフレームの冒頭5件を出力
df.head()
```

```python
# 'accesses'列のデータをNumPy配列として取得
data = df["国語"].values
# 平均値を出力
mean = np.mean(data)

# 最小値を求める
min = np.min(data)

# 最大値を求める
max = np.max(data)

# 中央値を求める
median = np.median(data)

q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)


# 平均値，中央値，最小値，最大値，第1四分位数，第3四分位数
print(f"平均値: {mean}")
print(f"中央値: {median}")
print(f"最小値: {min}")
print(f"最大値: {max}")
print(f"第1四分位数: {q1}")
print(f"第3四分位数: {q3}")
```
