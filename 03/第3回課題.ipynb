```python
#全ての説明変数を利用する場合
import pandas as pd
from sklearn.linear_model import LinearRegression

#読み込み
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/第3回/data4_7.csv', index_col=0)

#目的変数y、説明変数X
X = df.drop('GPA', axis=1)
y = df['GPA']

#重回帰モデルの作成
model = LinearRegression()
model.fit(X, y)

#出力
for col, coef in zip(X.columns, model.coef_):
    print(f"  {col}: {coef:.4f}")
```

```python
#ステップワイズ法
import pandas as pd
from sklearn.linear_model import LinearRegression
from mlxtend.feature_selection import SequentialFeatureSelector

#読み込み
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/第3回/data4_7.csv', index_col=0)
X = df.drop('GPA', axis=1)
y = df['GPA']

#ステップワイズ法
sfs = SequentialFeatureSelector(
    LinearRegression(),
    k_features=3,
    forward=True,
    scoring='r2',
    cv=0
)

sfs.fit(X, y)

stepwize_features = list(sfs.k_feature_names_)
print(f"ステップワイズで選択された変数: {stepwize_features}")

#抽出
X_selected = X[stepwize_features]

model_stepwise = LinearRegression()
model_stepwise.fit(X_selected, y)

#出力
for col, coef in zip(X_selected.columns, model_stepwise.coef_):
    print(f"{col}: {coef:.4f}")
```

```python
#サブセット法
import pandas as pd
from sklearn.linear_model import LinearRegression
from itertools import combinations

#読み込み
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/第3回/data4_7.csv', index_col=0)
X = df.drop('GPA', axis=1)
y = df['GPA']

best_r2, best_features = -1, None

for k in range(1, len(X.columns) + 1):
    for features in combinations(X.columns, k):
        r2 = LinearRegression().fit(X[list(features)], y).score(X[list(features)], y)
        if r2 > best_r2:
            best_r2, best_features = r2, features

print(f"サブセットで選択された変数: {list(best_features)}")

X_best = X[list(best_features)]
model_best = LinearRegression()
model_best.fit(X_best, y)

for col, coef in zip(X_best.columns, model_best.coef_):
    print(f"{col}: {coef:.4f}")

print(f"\n決定係数 (R^2 score): {best_r2:.4f}")
```

このデータ表に対して，目的変数を製品長さ，説明変数を材料重量，滞留時間，環境温度，加工機械とする重回帰分析を実施せよ．このとき，すべての変数を利用する場合，ステップワイズ法を適用する場合，サブセット法を適用する場合についての重回帰分析を，それぞれ実施せよ．

```python
#全ての変数の場合
import pandas as pd
from sklearn.linear_model import LinearRegression

# 読み込み
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/第3回/data4_8.csv')

#説明変数と目的変数
y = df['製品長さy']
X = df[['材料重量x1', '滞留時間x2', '環境温度x3', '加工機械x4']]

#重回帰モデルの作成
model = LinearRegression()
model.fit(X, y)

#出力
for col, coef in zip(X.columns, model.coef_):
    print(f"{col}: {coef:.4f}")

print(f"\n決定係数 (R^2 score): {model.score(X, y):.4f}")
```

```python
#ステップワイズ法
import pandas as pd
from sklearn.linear_model import LinearRegression
from mlxtend.feature_selection import SequentialFeatureSelector

#読み込み
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/第3回/data4_8.csv')
X = df[['材料重量x1', '滞留時間x2', '環境温度x3', '加工機械x4']]
y = df['製品長さy']

sfs = SequentialFeatureSelector(
    LinearRegression(),
    k_features=3,
    forward=True,
    scoring='r2',
    cv=0
)

sfs.fit(X, y)

stepwize_features = list(sfs.k_feature_names_)
print(f"ステップワイズで選択された変数: {stepwize_features}")

#抽出
X_selected = X[stepwize_features]

model_stepwise = LinearRegression()
model_stepwise.fit(X_selected, y)

#出力
for col, coef in zip(X_selected.columns, model_stepwise.coef_):
    print(f"{col}: {coef:.4f}")
```

```python
#サブセット法
import pandas as pd
from sklearn.linear_model import LinearRegression
from mlxtend.feature_selection import ExhaustiveFeatureSelector

#読み込み
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/第3回/data4_8.csv')

X = df[['材料重量x1', '滞留時間x2', '環境温度x3', '加工機械x4']]
y = df['製品長さy']

efs = ExhaustiveFeatureSelector(
    LinearRegression(),
    min_features=1,
    max_features=4,
    scoring='r2',
    cv=0
)

efs.fit(X, y)

subset_features = list(efs.best_feature_names_)
print(f"サブセット法で選択された変数: {subset_features}")

#抽出
X_selected = X[subset_features]

model_subset = LinearRegression()
model_subset.fit(X_selected, y)

#出力
for col, coef in zip(X_selected.columns, model_subset.coef_):
    print(f"{col}: {coef:.4f}")
```
