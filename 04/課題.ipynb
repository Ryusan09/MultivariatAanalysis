# 演習1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# データの読み込み
df = pd.read_csv('prob_6-5.csv')
print('データの形状:', df.shape)
df
```

```python
# データの基本統計量を確認
df.describe()
```

## (1) 説明変数を1つだけ使ったロジスティック回帰分析

```python
# 目的変数
y = df['y'].values

# 各説明変数ごとにロジスティック回帰を実行
# 正則化を実質的に無効化するため C を大きく設定
results_single = {}

for col in ['x1', 'x2', 'x3', 'x4']:
    X = df[[col]].values  # 説明変数（2次元配列にする）
    
    model = LogisticRegression(C=1e10, solver='lbfgs', max_iter=10000)
    model.fit(X, y)
    
    b0 = model.intercept_[0]
    b1 = model.coef_[0][0]
    accuracy = model.score(X, y)
    
    results_single[col] = {'b0': b0, 'b1': b1, 'accuracy': accuracy, 'model': model}
    
    print(f'--- 説明変数: {col} ---')
    print(f'  b_0 (切片) = {b0:.4f}')
    print(f'  b_1 (係数) = {b1:.4f}')
    print(f'  回帰式: p = 1 / (1 + exp(-({b0:.4f} + {b1:.4f} * {col})))')
    print(f'  学習データでの正答率: {accuracy:.4f}')
    print()
```

```python
# 結果をまとめた表
summary_single = pd.DataFrame({
    '説明変数': list(results_single.keys()),
    'b_0 (切片)': [results_single[c]['b0'] for c in results_single],
    'b_1 (係数)': [results_single[c]['b1'] for c in results_single],
    '正答率': [results_single[c]['accuracy'] for c in results_single],
})
summary_single
```

## (2) すべての説明変数を使ったロジスティック回帰分析

```python
# 4つの説明変数すべてを使ったロジスティック回帰
X_all = df[['x1', 'x2', 'x3', 'x4']].values
y = df['y'].values

model_all = LogisticRegression(C=1e10, solver='lbfgs', max_iter=10000)
model_all.fit(X_all, y)

b0_all = model_all.intercept_[0]
b_all = model_all.coef_[0]
accuracy_all = model_all.score(X_all, y)

print('=== すべての説明変数を用いたロジスティック回帰 ===')
print(f'  b_0 (切片) = {b0_all:.4f}')
print(f'  b_1 (x1の係数) = {b_all[0]:.4f}')
print(f'  b_2 (x2の係数) = {b_all[1]:.4f}')
print(f'  b_3 (x3の係数) = {b_all[2]:.4f}')
print(f'  b_4 (x4の係数) = {b_all[3]:.4f}')
print()
print(f'  回帰式: p = 1 / (1 + exp(-({b0_all:.4f} '
      f'+ {b_all[0]:.4f}*x1 + {b_all[1]:.4f}*x2 '
      f'+ {b_all[2]:.4f}*x3 + {b_all[3]:.4f}*x4)))')
print()
print(f'  学習データでの正答率: {accuracy_all:.4f}')
```
