```python
import numpy as np
import pandas as pd
from scipy.stats import zscore
import matplotlib.pyplot as plt
```

```python
my_data = pd.read_csv('prob_6-5.csv', encoding="shift-jis")
```

```python
my_data
```

```python
X1, y  = my_data["x1"], my_data["y"]
X1 = X1.values.reshape(-1, 1)

X2, y  = my_data["x2"], my_data["y"]
X2 = X2.values.reshape(-1, 1)

X3, y = my_data["x3"], my_data["y"]
X3 = X3.values.reshape(-1, 1)

X4, y = my_data["x4"], my_data["y"]
X4 = X4.values.reshape(-1, 1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X1, y)

print(model.coef_)
print(model.intercept_)
```

```python
model.fit(X2, y)

print(model.coef_)
print(model.intercept_)
```

```python
# 定数項
model.fit(X3, y)

print(model.coef_)
print(model.intercept_)
```

```python
model.fit(X4, y)

print(model.coef_)
print(model.intercept_)
```

```python
model.fit(X1,X2,X3,X4, y)

print(model.coef_)
print(model.intercept_)
```

 $b_0 = -6.000$，$b_1= 0.1929$ と求められた．
