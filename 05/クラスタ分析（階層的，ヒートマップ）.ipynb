```python
import pandas as pd
import seaborn as sns

my_data = pd.DataFrame(
    {'language': [  0,  20,  20,  25,  22,  17],
     'english':  [  0,  20,  40,  20,  24,  18],
     'math':     [100,  20,   5,  30,  17,  25],
     'science':  [  0,  20,   5,  25,  16,  23],
     'society':  [  0,  20,  30,   0,  21,  17]},
    index=       ['A', 'B', 'C', 'D', 'E', 'F'])
```

```python
my_data
```

科目毎に点数を標準化してからヒートマップを書く．

```python
sns.clustermap(my_data, z_score=1) # 列ごとの標準化
```

受験者の類似度を見ると，FとBが一番近く，Aが一番離れていることが分かる．実際，Aは極端な点数となっている．

受験者をつかって科目のデンドログラムによる類似度も見ることができる．
英語と社会がまとまっていることから類似度が高いことや，国語と理科のまとまりから類似度が高いことが読み取れる．
